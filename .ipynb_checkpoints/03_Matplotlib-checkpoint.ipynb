# Introducción a Matplotlib

[Matplotlib](http://matplotlib.org) es una biblioteca que permite la creación de figuras y gráficos de calidad mediante el uso de Python.
* Permite la creación de gráficos de manera sencilla y eficiente.
* Permite la integración de gráficos y figuras en un Jupyter Notebook

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Muestra los gráficos inegrados dentro de un Jupyter Notebook
%matplotlib inline
```

## Representación gráfica de Datos.

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como en la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7,],[1, 4, 6 , 7, 8])
plt.show
```

Puede modificarse las longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3, -1, 0, 4, 7,],[1, 4, 6 , 7, 8])
plt.axis([-4, 8, 0, 10]) # [x = min, x = max, y = min, y  = max]
plt.show()
```

Se sigue el mismo procedimineto para pintar una función matemática.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x,y)
plt.show()
```

También puede modificarse el estudio de la gráfica para que contenga más información.

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las función

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

y2 = x + 1

plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")

plt.plot(x,y, 'b--', x, y2, 'g')
plt.grid(True)
plt.show()
```

```python
# separando en difetentes lineas las funciones
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x,y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1


plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x,y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") # la situa en la mejor localización
plt.show()
```

También puede crearse dos gráficas que no se superpongan. Estas gráficas se organizan en un grid y se denominan subplots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 Rows, 2 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')

plt.show()
```

Para que las gráficas no queden tan ajustadas se puede hacer la figura más grandes

```python
plt.figure(figzise = (14, 6))
plt.show
```
