## Representación gráfica de datos

Si la función de trazado se le da un matriz de datos, la usará como coordenadas en el eje vectorial y utilizará el índice de cada punto de datos en el array como la coordenada horitzontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal y otra para el eje vertical

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

```python
# Pueden modificarse los longitudes de los ejes para que la figura no se vea tan fit

plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [xmin, xmax, ymin, ymax]
plt.show()
```

Se sigue el msimo procedimiento para pintar una función matemática

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x, y)
plt.show()
```

También puede modificarse el estilo de la gráfica para que contenga más información

```python
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
y2= x + 1
plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

Separar en diferentes líneas las funciones

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.title("Doble Función")
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda.

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.title("Dable Función")
plt.xlabel("x")
plt.ylabel("y = x², y2 = x + 1")
plt.grid("True")

plt.plot(x, y, 'b--', label = "x²")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") # Lo situa en la mejor localización
plt.show()
```

También se pueden crear dos gráficas que no se superpongan. Estás gráficas se organizan en un grid y se denominan subplots

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 rows, 2 columns, 1st subplot
plt.title("Square Función")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid("True")
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.title("Lineal Función")
plt.xlabel("x")
plt.ylabel("y2 = x + 1")
plt.grid("True")
plt.plot(x, y2, 'g')

plt.show()
```

para que las gráficas no queden tan ajustadas es posible hacer la figura más grande.

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1) # 1 rows, 2 columns, 1st subplot
plt.title("Square Función")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y = x²", fontsize = 14)
plt.grid("True")
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.title("Lineal Función")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y2 = x + 1", fontsize = 14)
plt.grid("True")
plt.plot(x, y2, 'g')

plt.show()
```

## Scatter Plots

```python
from numpy.random import rand

x, y = rand(2, 100)
plt.scatter(x, y)
plt.title("Scatter Plots")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.grid("True")
plt.show()
```

```python
from numpy.random import rand

x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.title("Scatter Plots")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.grid("True")
plt.show()
```

## Histogramas

```python
data = [1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3,]
plt.subplot(2, 1, 1,) # 2 rows, 1 columns, 1st subplot
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(2, 1, 2) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel('Value')
plt.ylabel('Frecuency')

plt.show()
```

## Guardar los gráficos

```python
from numpy.random import rand

x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.title("Scatter Plots")
plt.xlabel("x", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.grid("True")
plt.savefig("Grafico_de_dispersion.", transparent = True)
```
